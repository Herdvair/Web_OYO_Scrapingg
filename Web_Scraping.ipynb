## Data Scraping Menggunakan BeautifulSoup

Dilakukan web scraping di platform OYO dengan mengambil tanggal 7-14 Juli 2025

```python
# API URL
url_hotel = 'https://www.oyorooms.com/search?location=Makassar%2C%20South%20Sulawesi%2C%20Indonesia&city=Makassar&searchType=city&coupon=&checkin=07%2F07%2F2025&checkout=14%2F07%2F2025&roomConfig%5B%5D=2&showSearchElements=false&guests=2&rooms=1&countryName=Indonesia&country=indonesia&filters%5Bcity_id%5D=700'
option = webdriver.ChromeOptions()
option.add_argument("--start-maximized")

# uncomment untuk running code dibawah ini jika run di google collab
option.add_argument("--headless")
option.add_argument("--disable-gpu")
option.add_argument("--no-sandbox")
option.add_argument("--remote-debugging-port=9230")
# comment 4 code diatas (option.add) jika running di local / jupyter notebook dsb
driver_hotel = webdriver.Chrome(options=option)
driver_hotel.get(url_hotel)
# Membuat objek BeautifulSoup dari konten HTML
soup = BeautifulSoup(driver_hotel.page_source, "html.parser")
```

```python
for tag in soup.find_all(True):
    print(tag.name)
```

```python
soup
```

```python
div_containers_oyo = soup.find_all("div", attrs={'class' :'oyo-cell--12-col oyo-cell--8-col-tablet oyo-cell--4-col-phone'})
print(div_containers_oyo)
```

### One Page (karena OYO hanya menampilkan satu halaman untuk informasi akomodasinya)

1. Nama Hotel

```python
hotel_oyo = []
for container in div_containers_oyo:
    list_nama_oyo = container.find('h3', attrs={"class":'listingHotelDescription__hotelName d-textEllipsis'})
    if list_nama_oyo:
      nama_oyo = list_nama_oyo.text.strip()
    else:
      nama_oyo = None
    hotel_oyo.append(nama_oyo)

hotel_oyo = pd.DataFrame(hotel_oyo, columns=['Name'])
hotel_oyo
```

2. Harga

```python
harga_hotel_oyo = []
for container in div_containers_oyo:
    list_harga_oyo = container.find('span', attrs={"class":'listingPrice__finalPrice listingPrice__finalPrice--black'})
    if list_harga_oyo:
      harga_oyo = list_harga_oyo.text.strip()
    else:
      harga_oyo = None
    harga_hotel_oyo.append(harga_oyo)

harga_hotel_oyo = pd.DataFrame(harga_hotel_oyo, columns=['Price'])
harga_hotel_oyo
```

3. Lokasi

```python
lokasi_hotel_oyo = []
for container in div_containers_oyo:
    list_lokasi_oyo = container.find('span', attrs={"class":'u-line--clamp-2'})
    if list_lokasi_oyo:
      lokasi_oyo = list_lokasi_oyo.text.strip()
    else:
      lokasi_oyo = None
    lokasi_hotel_oyo.append(lokasi_oyo)

lokasi_hotel_oyo = pd.DataFrame(lokasi_hotel_oyo, columns=['Location'])
lokasi_hotel_oyo
```

4. Wizard Member

```python
member_hotel_oyo = []
for container in div_containers_oyo:
    wizard_member_oyo = container.find('div', attrs={"class":'oyo-cell--12-col listingHotelDescription__labels'})
    if wizard_member_oyo:
      member_oyo = wizard_member_oyo.text.strip()
    else:
      member_oyo = None
    member_hotel_oyo.append(member_oyo)

member_hotel_oyo = pd.DataFrame(member_hotel_oyo, columns=['Membership'])
member_hotel_oyo
```

5. Rating

```python
rating_hotel_oyo = []
for container in div_containers_oyo:
    list_rating_oyo = container.find('div', attrs={"class":'hotelRating__wrapper'})
    if list_rating_oyo:
      rating_oyo = list_rating_oyo.text.strip()
    else:
      rating_oyo = None
    rating_hotel_oyo.append(rating_oyo)

rating_hotel_oyo = pd.DataFrame(rating_hotel_oyo, columns=['Rating'])
rating_hotel_oyo
```

### Data Cleaning

Melakukan Kombinasi dari berbagai kolom yaitu nama hotel, harga, lokasi, membership, rating)

```python
df_oyo = pd.concat([hotel_oyo,harga_hotel_oyo, lokasi_hotel_oyo, member_hotel_oyo, rating_hotel_oyo], axis=1)

# print total hotel
print('Total Hotels : ', len(df_oyo))

df_oyo
```

**Deskripsi**

*   Saya menggunakan fungsi pd.concat() dari library Pandas untuk menggabungkan beberapa dataframe menjadi satu dataframe dengan nama df_oyo. Parameter axis=1 menunjukkan bahwa penggabungan ini dilakukan secara horizontal.
*   Menggunakan fungsi len() untuk mengetahui jumlah hotel yang berhasil di scraping

**Dampak dari integritas data**


*  Penggabungan ini penting untuk memastikan bahwa semua informasi terkait hotel tersedia di satu tempat. Namun, jika terdapat ketidakcocokan jumlah baris antara DataFrame yang digabungkan, terdapat resiko data tidak sinkron yang akan membuat baris kosong (NaN) di kolom yang tidak punya nilai.
